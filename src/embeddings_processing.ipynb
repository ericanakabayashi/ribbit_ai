```python
import os
import warnings
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from IPython.display import Audio
import soundfile as sf

warnings.filterwarnings("ignore", category=UserWarning)

class AudioPipeline:
    """
    AudioPipeline that works with both S3 and local/Google Drive files. Asked Claude to fix my ap.py file from capstone, in order to adapt it to google drive / colab setup.
    """

    def __init__(self, file_key=None, file_path=None, s3_bucket=None,
                 method='STFT', spect='mel', remove_background=True,
                 remove_crop_audio=True, plot_spectrogram=True,
                 plot_envelope=True, normalize_audio=False, sr=48000,
                 add_augmentation=False, bg_bucket=None, bg_file_key=None,
                 bg_file_path=None, play_finalaudio=False):

        # Support both S3 and local files
        self.file_key = file_key
        self.file_path = file_path
        self.s3_bucket = s3_bucket
        self.use_s3 = s3_bucket is not None

        self.sampling_rate = sr
        self.spect = spect
        self.audio_data = None
        self.cleaned_audio = None
        self.spectrogram = None
        self.envelope = None
        self.bg_noise = None

        self.remove_background = remove_background
        self.method = method
        self.remove_crop_audio = remove_crop_audio
        self.plot_spectrogram = plot_spectrogram
        self.plot_envelope = plot_envelope
        self.normalize_audio = normalize_audio
        self.add_augmentation = add_augmentation

        self.bg_bucket = bg_bucket
        self.bg_file_key = bg_file_key
        self.bg_file_path = bg_file_path
        self.play_finalaudio = play_finalaudio

    def load_audio_from_local(self):
        """Load audio from local file (Google Drive)."""
        try:
            if not self.file_path:
                raise ValueError("file_path is not set")

            if not os.path.exists(self.file_path):
                raise FileNotFoundError(f"File not found: {self.file_path}")

            duration = 3 if self.remove_crop_audio else None
            self.audio_data, self.sampling_rate = librosa.load(
                self.file_path,
                sr=self.sampling_rate,
                duration=duration
            )

            if self.normalize_audio:
                self.gaussian_norm_audio()

            if self.audio_data is None or len(self.audio_data) == 0:
                raise ValueError("Loaded audio data is empty.")

        except Exception as e:
            print(f"Error loading audio from local file: {e}")
            raise

    def load_audio_from_s3(self):
        """Load the audio from S3 bucket."""
        if self.s3_bucket and self.file_key:
            try:
                import boto3
                import io

                s3 = boto3.client('s3')
                obj = s3.get_object(Bucket=self.s3_bucket, Key=self.file_key)
                audio_data = obj['Body'].read()

                duration = 3 if self.remove_crop_audio else None
                self.audio_data, self.sampling_rate = librosa.load(
                    io.BytesIO(audio_data),
                    sr=self.sampling_rate,
                    duration=duration
                )

                if self.normalize_audio:
                    self.gaussian_norm_audio()

                if self.audio_data is None or len(self.audio_data) == 0:
                    raise ValueError("Loaded audio data is empty.")

            except Exception as e:
                print(f"Error loading audio from S3: {e}")
                raise

    def load_audio(self):
        """Load audio from either S3 or local file."""
        if self.use_s3:
            self.load_audio_from_s3()
        else:
            self.load_audio_from_local()

    def gaussian_norm_audio(self):
        """Apply Gaussian normalization to the audio data."""
        if self.audio_data is not None:
            mean = np.mean(self.audio_data)
            std = np.std(self.audio_data)
            if std > 0:
                self.audio_data = (self.audio_data - mean) / std
            else:
                print("Warning: Standard deviation is zero. Audio cannot be normalized.")
        else:
            print("Audio data is not available for normalization.")

    def load_background_audio(self):
        """Load background audio from S3 or local file."""
        try:
            if self.bg_file_path:
                if not os.path.exists(self.bg_file_path):
                    raise FileNotFoundError(f"Background file not found: {self.bg_file_path}")

                self.bg_noise, bg_sr = librosa.load(self.bg_file_path, sr=None)

            elif self.bg_bucket and self.bg_file_key:
                import boto3
                import io

                s3 = boto3.client('s3')
                obj = s3.get_object(Bucket=self.bg_bucket, Key=self.bg_file_key)
                bg_audio_data = obj['Body'].read()
                self.bg_noise, bg_sr = librosa.load(io.BytesIO(bg_audio_data), sr=None)
            else:
                print("No background audio file specified")
                return

            if bg_sr != self.sampling_rate:
                self.bg_noise = librosa.resample(
                    self.bg_noise,
                    orig_sr=bg_sr,
                    target_sr=self.sampling_rate
                )

            # Match length
            if len(self.bg_noise) < len(self.audio_data):
                self.bg_noise = np.pad(
                    self.bg_noise,
                    (0, len(self.audio_data) - len(self.bg_noise)),
                    'constant'
                )
            else:
                self.bg_noise = self.bg_noise[:len(self.audio_data)]

        except Exception as e:
            print(f"Error loading background audio: {e}")

    def add_background_noise(self):
        """Add background noise to the audio signal."""
        if self.bg_noise is not None and self.audio_data is not None:
            if not np.isfinite(self.bg_noise).all():
                print("Background noise contains non-finite values, skipping addition.")
                return

            max_bg = np.max(np.abs(self.bg_noise))
            if max_bg > 0:
                background_signal = self.bg_noise / max_bg
            else:
                print("Background noise max value is zero, skipping normalization.")
                background_signal = self.bg_noise

            min_length = min(len(self.audio_data), len(background_signal))
            self.audio_data[:min_length] += background_signal[:min_length]

    def remove_background_noise(self):
        """Remove background noise from the audio using the selected method."""
        if self.audio_data is not None:
            if self.method == 'STFT':
                self.cleaned_audio = self.remove_background_stft()
            elif self.method == 'PCEN':
                self.cleaned_audio = self.remove_background_pcen()
            elif self.method == 'HPF':
                self.cleaned_audio = self.apply_high_pass_filter(self.audio_data)

    def remove_background_stft(self):
        """Remove background noise using STFT."""
        if self.audio_data is not None:
            if not np.isfinite(self.audio_data).all():
                raise ValueError("Audio data contains non-finite values.")

        stft_result = librosa.stft(self.audio_data)
        magnitude, phase = np.abs(stft_result), np.angle(stft_result)
        noise_profile = np.median(magnitude, axis=1, keepdims=True)
        cleaned_magnitude = np.where(magnitude < noise_profile, 0, magnitude)
        cleaned_stft = cleaned_magnitude * np.exp(1j * phase)

        return librosa.istft(cleaned_stft, dtype=np.float64)

    def remove_background_pcen(self):
        """Remove background noise using PCEN."""
        stft_result = librosa.stft(self.audio_data)
        magnitude = np.abs(stft_result)
        pcen_result = librosa.pcen(
            magnitude * (2**31),
            sr=self.sampling_rate,
            gain=1.1,
            hop_length=512,
            bias=2,
            power=0.5,
            time_constant=0.8,
            eps=1e-6,
            max_size=2
        )

        cleaned_stft = pcen_result * np.exp(1j * np.angle(stft_result))
        return librosa.istft(cleaned_stft, dtype=np.float64)

    def apply_high_pass_filter(self, y):
        """Apply a high-pass filter to the audio."""
        b, a = signal.butter(10, 2000 / (self.sampling_rate / 2), btype='highpass')
        return signal.lfilter(b, a, y)

    def compute_spectrogram(self):
        """Compute the FFT spectrogram."""
        audio_to_process = self.cleaned_audio if self.remove_background else self.audio_data
        stft_result = librosa.stft(audio_to_process)
        self.spectrogram = librosa.amplitude_to_db(np.abs(stft_result), ref=np.max)
        self.envelope = np.abs(self.spectrogram).max(axis=0)

    def compute_melspectrogram(self):
        """Compute the Mel spectrogram."""
        audio_to_process = self.cleaned_audio if self.remove_background else self.audio_data
        mel_spectrogram = librosa.feature.melspectrogram(y=audio_to_process, sr=self.sampling_rate)
        self.spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)
        self.envelope = np.abs(self.spectrogram).max(axis=0)

    def plot_results(self):
        """Plot the spectrogram and/or envelope."""
        num_plots = int(self.plot_spectrogram) + int(self.plot_envelope)
        if num_plots == 0:
            return

        plt.figure(figsize=(5.11, 0.96), dpi=100)

        if self.plot_spectrogram:
            plt.subplot(num_plots, 1, 1)
            plt.imshow(self.spectrogram, aspect='auto', origin='lower', cmap='gray')
            plt.axis('off')

        if self.plot_envelope:
            plt.subplot(num_plots, 1, 2 if self.plot_spectrogram else 1)
            plt.plot(self.envelope, color='gray')
            plt.axis('off')

        plt.tight_layout()
        plt.show()

    def play_audio(self, audio_data=None):
        """Play the final processed audio."""
        if audio_data is None:
            audio_data = self.cleaned_audio if self.cleaned_audio is not None else self.audio_data

        if audio_data is not None:
            return Audio(audio_data, rate=self.sampling_rate)
        else:
            print("No audio data available to play.")

    def run_pipeline(self):
        """Run the entire audio processing pipeline."""
        try:
            # Load audio
            self.load_audio()

            # Add background noise if requested
            if self.add_augmentation:
                self.load_background_audio()
                self.add_background_noise()

            # Remove background noise if requested
            if self.remove_background:
                self.remove_background_noise()

            # Compute spectrogram
            if self.spect == 'mel':
                self.compute_melspectrogram()
            else:
                self.compute_spectrogram()

            # Plot results
            self.plot_results()

            # Play audio if requested
            if self.play_finalaudio:
                audio = self.play_audio()
                from IPython.display import display
                display(audio)

        except Exception as e:
            print(f"Error in the audio processing pipeline: {e}")
            raise
```

```python
def extract_embeddings(audio_data, sample_rate=48000):
    audio = np.array(audio_data)
    if len(audio) < sample_rate * 3:
        pad_length = (sample_rate * 3) - len(audio)
        audio = np.pad(audio, (0, pad_length), 'constant')
    else:
        audio = audio[:sample_rate * 3]

    audio_input = np.expand_dims(audio, axis=0)
    embedding_vector = embeddings(audio_input)
    return embedding_vector.flatten()
```

```python
from google.colab import drive
drive.mount('/content/drive')

import chromadb
from chromadb.config import Settings
import numpy as np
import pandas as pd
from datetime import datetime
import os

CHROMA_DB_PATH = '/content/drive/MyDrive/ribbit/chromadb'
os.makedirs(CHROMA_DB_PATH, exist_ok=True)

client = chromadb.PersistentClient(
    path=CHROMA_DB_PATH,
    settings=Settings(
        anonymized_telemetry=False,
        allow_reset=True
    )
)

print(f"ChromaDB initialized at: {CHROMA_DB_PATH}")
collection = client.get_or_create_collection(
    name="anuraset_singlecall",
    metadata={
        "description": "BirdNET embeddings",
        "embedding_model": "BirdNET-Analyzer-V2.4",
        "embedding_dim": 1024,
        "created_at": datetime.now().isoformat()
    }
)

print(f"Collection '{collection.name}' ready")
print(f"Current size: {collection.count()} embeddings")

def add_embeddings_to_chromadb(
    embeddings,
    metadata_df,
    collection,
    batch_size=1000
):
    n_samples = len(embeddings)

    print(f"Adding {n_samples} embeddings to ChromaDB...")

    for i in range(0, n_samples, batch_size):
        batch_end = min(i + batch_size, n_samples)
        batch_embeddings = embeddings[i:batch_end]
        batch_metadata = metadata_df.iloc[i:batch_end]

        ids = [f"emb_{i+j:06d}" for j in range(len(batch_embeddings))]
        metadatas = []
        for _, row in batch_metadata.iterrows():
            meta = {
                'fname': str(row['fname']),
                'species': str(row['species']),
                'site': str(row['site']),
#                'species_number': int(row['species_number']),
            }

            if 'date' in row:
                meta['date'] = str(row['date'])
            if 'min_t' in row:
                meta['min_t'] = float(row['min_t'])
            if 'max_t' in row:
                meta['max_t'] = float(row['max_t'])

            metadatas.append(meta)

        collection.add(
            ids=ids,
            embeddings=batch_embeddings.tolist(),
            metadatas=metadatas
        )

        print(f"  Added batch {i//batch_size + 1}: {i+1}-{batch_end}/{n_samples}")

    print(f"Total embeddings in collection: {collection.count()}")


def query_similar_embeddings(
    query_embedding,
    collection,
    n_results=10,
    filter_species=None,
    filter_site=None
):
    where_filter = {}
    if filter_species:
        where_filter['species'] = filter_species
    if filter_site:
        where_filter['site'] = filter_site

    results = collection.query(
        query_embeddings=[query_embedding.tolist()],
        n_results=n_results,
        where=where_filter if where_filter else None
    )

    return results


def get_embeddings_by_species(collection, species, limit=None):
    results = collection.get(
        where={'species': species},
        limit=limit
    )
    return results


def export_chromadb_to_numpy(collection, output_dir):
    results = collection.get(
        include=['embeddings', 'metadatas']
    )
    embeddings = np.array(results['embeddings'])
    metadata_df = pd.DataFrame(results['metadatas'])

    os.makedirs(output_dir, exist_ok=True)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M")

    emb_path = os.path.join(output_dir, f'{timestamp}_chromadb_embeddings.npy')
    meta_path = os.path.join(output_dir, f'{timestamp}_chromadb_metadata.csv')

    np.save(emb_path, embeddings)
    metadata_df.to_csv(meta_path, index=False)

    print(f" Exported {len(embeddings)} embeddings")
    print(f"  Embeddings: {emb_path}")
    print(f"  Metadata: {meta_path}")

    return embeddings, metadata_df


def process_and_store_embeddings(
    file_paths,
    metadata_df,
    collection,
    sr=48000,
    batch_size=100,
    chromadb_batch_size=1000
):
    from tqdm.auto import tqdm

    total_processed = 0
    batch_embeddings = []
    batch_metadata = []
    failed_files = []

    print(f"Processing {len(file_paths)} files...")
    print(f"Will store to ChromaDB every {chromadb_batch_size} embeddings\n")

    for i, file_path in enumerate(tqdm(file_paths)):
        try:
            audio_data = load_and_process_audio(file_path, sr=sr)
            embedding = extract_embeddings(audio_data, sample_rate=sr)
            batch_embeddings.append(embedding)
            batch_metadata.append(metadata_df.iloc[i])

            if len(batch_embeddings) >= chromadb_batch_size:
                batch_df = pd.DataFrame(batch_metadata)
                add_embeddings_to_chromadb(
                    np.array(batch_embeddings),
                    batch_df,
                    collection,
                    batch_size=chromadb_batch_size
                )

                total_processed += len(batch_embeddings)
                batch_embeddings = []
                batch_metadata = []

        except Exception as e:
            failed_files.append((i, file_path, str(e)))
            continue

    if batch_embeddings:
        batch_df = pd.DataFrame(batch_metadata)
        add_embeddings_to_chromadb(
            np.array(batch_embeddings),
            batch_df,
            collection,
            batch_size=len(batch_embeddings)
        )
        total_processed += len(batch_embeddings)

    print(f"\n Processing complete!")
    print(f"  Total stored: {total_processed}")
    print(f"  Failed: {len(failed_files)}")
    print(f"  Collection size: {collection.count()}")

    return failed_files


print("\n" + "="*60)
print(" ChromaDB setup complete!")
print("="*60)
```

```python
TEST_RUN = False
test_size = 5

df = singlespecies.copy()
if TEST_RUN:
    df = df.head(test_size)

print(f"Processing {len(df)} files...")

failed_files = []
batch_embeddings = []
batch_metadata = []

for i, row in tqdm(df.iterrows(), total=len(df)):
    file_path = f"{AUDIO_ROOT}/{row['site']}/{row['filename']}"

    try:
        pipeline = AudioPipeline(
            file_path=file_path,
            method="STFT",
            spect='mel',
            remove_background=True,
            remove_crop_audio=True,
            plot_spectrogram=False,
            plot_envelope=False,
            normalize_audio=False,
            sr=48000,
            add_augmentation=False,
            bg_file_path=None,
            play_finalaudio=False
        )
        pipeline.run_pipeline()

        embedding = extract_embeddings(pipeline.audio_data, sample_rate=48000)
        batch_embeddings.append(embedding)
        batch_metadata.append(row)

    except Exception as e:
        print(f"  ✗ Failed [{i}] {file_path}: {e}")
        failed_files.append((i, file_path, str(e)))
        continue

# push to chroma unsing pre-defined functions
if batch_embeddings:
    batch_df = pd.DataFrame(batch_metadata)
    add_embeddings_to_chromadb(
        np.array(batch_embeddings),
        batch_df,
        collection,
        batch_size=len(batch_embeddings)
    )

print(f"\nStored: {len(batch_embeddings)} | Failed: {len(failed_files)}")
print(f"Collection size: {collection.count()}")
```
